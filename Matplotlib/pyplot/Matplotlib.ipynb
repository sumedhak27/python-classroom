```python
import matplotlib.pyplot as plt
import numpy as np
```

# Plot 

```python
days = [1,2,3,4,5,6,7]
youtube = [23,44,12,2,87,34,23]
facebook = [34,23,14,27,52,23,34]
plt.title('Video Performance')
plt.xlabel('days')
plt.ylabel('views')
plt.plot(days,youtube,'r*-', label = "YouTube")
plt.plot(days,facebook,'bo-', label = "facebook")
plt.legend(loc = 'upper left')
plt.ylim(0, 100)
plt.grid(True, color = "#efefef", linewidth = 2, linestyle= "--")
plt.savefig('F:\Python Classroom\Matplotlib\pyplot\img1.png', dpi = 500)
plt.show()
```

# Scatter plot

```python
plt.title('Video Performance')
plt.xlabel('days')
plt.ylabel('views')
plt.scatter(days,youtube,color = 'red', label = "YouTube", linewidth = 2)
plt.scatter(days,facebook,color = 'blue', label = "facebook")
plt.legend(loc = 'upper left')
plt.ylim(0, 100)
plt.grid(True, color = "#efefef", linewidth = 2, linestyle= "--", alpha = 0.5)
plt.savefig('F:\Python Classroom\Matplotlib\pyplot\img2.png', dpi = 500)
plt.show()
```

# Bar Graph

```python
plt.title('Video Performance')
plt.xlabel('Days')
plt.ylabel('Views')
plt.grid(True, color = "#efefef", linewidth = 1, linestyle= "--", alpha = 0.5)
plt.bar([a-0.25 for a in days], youtube, color = 'red', width = 0.25,label = "YouTube", align='edge')
plt.bar([a+0.25 for a in days], facebook, color = 'blue',width =-0.25, label = "facebook", align='edge')
plt.legend(loc = 'upper left')
plt.ylim(0, 100)
# plt.xticks([a+2 for a in days])
plt.savefig('F:\Python Classroom\Matplotlib\pyplot\img3.png', dpi = 700)
plt.show()
```

# Histogram

```python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# the histogram of the data
n, bins, patches = plt.hist(x, 50, density=1, facecolor='g', alpha=0.75)


plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(50, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```
